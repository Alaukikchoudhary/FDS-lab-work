# STEM PLOT

```python
import matplotlib.pyplot as plt
x = [1, 3, 6, 8]
y = [2, 4, 5, 7]
plt.stem(x,y,linefmt=":",markerfmt="r+",bottom=0,basefmt="g",label="python",use_line_collection=False)
plt.legend()
plt.title("stem plot")
plt.show()
```

# Horizontal steam plot

```python
import matplotlib.pyplot as plt
x = [1, 3, 6, 8]
y = [2, 4, 5, 7]
plt.stem(x,y,linefmt=":",markerfmt="r+",bottom=0,basefmt="g",label="python",use_line_collection=True,orientation="horizontal")
plt.legend()
plt.title("stem plot")
plt.show()
```
